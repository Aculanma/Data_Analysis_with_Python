### 1. Underflow error

```python
our_number = 1e-12
x = 1e-1
while our_number + x != our_number:
    x /= 10
print(x*10)
```

### 2. Управление пекарней

```python
import math


def bakery_products(day: int):
    return int(math.exp(math.sin(day/100) * day/10) * 20)
```

```python
bakery_products(18)
```

### 3. Битва при Азенкуре 2.0

```python
with open('input.txt', 'r') as file:
        dates = file.read().split()

def bin_search(lst):
    our_date = 1415
    low = 0
    high = len(lst) - 1

    while low <= high:
        mid = (low + high) // 2
        guess = int(lst[mid])
        if guess == our_date:
            return mid
        if guess > our_date:
            high = mid -1
        else:
            low = mid + 1
    return 'Элемент не найден'
    
```

```python
bin_search(dates)
```

### 4. Менеджмент

```python
import numpy as np


with open('input.txt', 'r') as file:
    data = file.read().split()
    
    plan = np.linspace(float(data[0]), float(data[1]), int(data[2]))
    plan[::7] = plan[::7]/ 3
    plan[4::7] = plan[4::7] * 2
    with open('output.txt', 'w') as file:
        for num in plan:
            print(f'{num:.2f}', file=file)
```
