# ДЗ 3. Итерируемые объекты

### 1. Задом наперед

```python
file = "input.txt"
with open(file, "r") as file:
    text = file.read()
for i in text.split():
    print(i[::-1], end=' ')
```

### 2. Буква ё

```python
file = "input.txt"
with open(file, "r") as file:
    text = file.readlines()

filtered = filter(lambda line: 'ё' in line, text)
changed_and_filtered = map(lambda s: s.strip(), filtered)
for i in changed_and_filtered:
    print(i)
```

### 3. Убрать половину

```python
file = "input.txt"
with open(file, "r") as file:
    text = file.read()
    text = text.splitlines()

for line in text:
    line = line.split()[::2]
    print(' '.join(line))
```

### 4. Дни рождения

__Первый способ__ решения:

```python
with open('input.txt', "r") as file:
    text = file.read()
    text = text.splitlines()
    
text = text[1:]
a = text[:len(text)//2]
b = text[len(text)//2:]

with open('output.txt', 'w') as f:
    for name, date in zip(a, b):
        print(name, date, sep='\t', file=f)
```

__Второй способ__ решения

```python
with open('input.txt', 'r') as file:
    text = file.read().split('\n')[1:]
    a = text[:len(text)//2]
    b = text[len(text)//2:]
    with open('output.txt', 'w') as f:
        f.write('\n'.join(list(map(lambda x: x[0] + '\t' + x[1], zip(a,b)))))  
```

### 5. Комбинации дежурных

```python
import itertools as it


with open('studygroup.txt', 'r') as file:
    text = file.read().split()
for x, y, z in it.combinations(text, 3):
    print('1:', x, '2:', y, '3:', z)
```
