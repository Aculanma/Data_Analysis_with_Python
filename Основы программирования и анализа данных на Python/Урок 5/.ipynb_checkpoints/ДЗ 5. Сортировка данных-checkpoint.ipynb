# ДЗ 5. Сортировка данных

### 1. Две команды

```python
with open('weights.txt', 'r') as file:
    text = file.read().splitlines()
    text.sort(key = lambda x: float(x.split()[1]))
    
    team_1 = text[1::2][::-1]
    team_2 = text[::2][::-1]

    with open ('team.txt', 'w') as file:
        for i in team_1:
            print(i, file=file)
        for j in team_2:
            print(j, file=file)
```

### 2. Неопубликованный шедевр

```python
with open('poe_unpublished.txt', 'r') as file:
    text = file.read().splitlines()
    text.sort(key = lambda x: len(x.split()))

    with open ('poe_decode_attempt.txt', 'w') as file:
        for i in text:
            new_line = i.split()
            new_line.sort(key = lambda x: len(x))
            print(' '.join(new_line), file=file)
```

### 3. Транспонирование

```python
with open('med_research.txt', 'r') as file:
    lines = file.read().splitlines()
    lines = [line.split(' ') for line in lines]

    with open ('output.txt', 'w') as file:
        columns = len(lines[0])
        for i in range(columns):
            print(*[line[i] for line in lines], file = file)
```

### 4. Логирование звонков

```python
with open('the_calls.txt', 'r') as file:
    lines = file.read().splitlines()
    
    assistant_A = filter(lambda line: line.split('\t')[2] == 'A', lines)
    assistant_B = filter(lambda line: line.split('\t')[2] == 'B', lines)
    
    assistant_A = sorted(assistant_A, key = lambda line: float(line.split('\t')[1]), reverse=True)
    assistant_B = sorted(assistant_B, key = lambda line: float(line.split('\t')[1]), reverse=True)
    
    with open('calls.txt', 'w') as file:
        print(*assistant_A, sep='\n', file = file)
        print(*assistant_B, sep='\n', file = file, end='')
```
