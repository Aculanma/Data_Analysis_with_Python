# ДЗ 8. Работа с библиотекой NumPy

### 1. Немного статистики

```python
import numpy as np


with open('input.txt', 'r') as file:
    nums = file.read().split()
    num = [int(num) for num in nums]
    print(f'{np.median(num):.2f} {np.mean(num):.2f} {np.std(num):.2f}')
```

### 2. Мода

```python
import numpy as np


matrix = np.genfromtxt('input.csv', delimiter=',')
i = np.sum(matrix[:,:1])
j = np.sum(matrix[:,1:2])
k = np.sum(matrix[:,2:3])

maximum = max(i, j, k)
if maximum == i:
    print(1)
elif maximum == j:
    print(2)
else:
    print(3)
```

### 3. Две чугунные гири

```python
import numpy as np


matrix = np.genfromtxt('input.csv', delimiter=',')
first = 0
second = 0
for i in matrix:
    if np.std(i) > 4:
        second += 1
    else:
        first += 1
if first > second:
    print(1)
else:
    print(2)
```

### 4. Очистка логов

```python
import numpy as np


matrix = np.genfromtxt('input.csv', delimiter=',')
for i in range(30):
    for j in range(12):
        if (i+j)%2!=0:
            matrix[i][j] = matrix[i][j]/2
np.savetxt('output.csv', matrix,fmt='%g', delimiter=',')
```

### 5. Дуэльный клуб

```python
import numpy as np


matrix = np.genfromtxt('input.csv', delimiter=',')
for i in range(30):
    avg = np.mean(matrix[:,i])
    matrix[0,i] = int(1.5 * avg)
np.savetxt('output.csv', matrix, fmt='%g', delimiter=',')
```
