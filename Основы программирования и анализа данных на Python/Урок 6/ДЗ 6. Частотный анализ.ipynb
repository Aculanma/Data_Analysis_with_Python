### 4. Наиболее часто встречающаяся цифра

```python
x = input()
dict_digits = {} 
for digit in x:
    if digit not in dict_digits:
        dict_digits[digit] = 1
    else:
        dict_digits[digit] += 1
rating = sorted(dict_digits.items(), key = lambda pair: pair[1], reverse=True)
print(rating[0][0])
```

### 5. Все виды животных

```python
with open('input.txt', 'r') as file:
    text = file.readlines()
    
    animals = set(line.split()[1] for line in text)
    animals_list = list(animals)
    animals_list.sort(key = len, reverse=True)
    print(*animals_list, sep='\n')
```

### 6. Создание потомства

```python
with open('input.txt', 'r') as file:
    text = file.readlines()
    
    male = {line.split()[1] for line in text if line.split()[2] == 'male'}
    female = {line.split()[1] for line in text if line.split()[2] == 'female'}

    animals = list(male & female)
    animals.sort(key=len)
    if len(animals) != 0:
        print(*animals, sep='\n')
    else:
        print(0)
```

### 7. Наиболее часто встречающееся животное

```python
from collections import Counter


with open('input.txt', 'r') as file:
    text = file.readlines()

    animal = [line.split()[1] for line in text]
    count_animals = Counter(animal).most_common()
    for animal, count in count_animals:
        print(f'{animal} - {count}') 
```

### 8. ID животных*

```python
from collections import defaultdict


with open('input.txt', 'r') as file:
    animals = defaultdict(list)
    
    for line in file:
        id, name = line.split(' ')[:2]
        animals[name].append(id)

    for name in sorted(animals.keys(), key = lambda a: len(a)):
        print("{}: {}".format(name, ', '.join(sorted(animals[name], key = lambda a: int(a)))))
```
