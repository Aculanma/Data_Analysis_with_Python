# ДЗ 4. Числовые данные

### 1. Перевод из двоичной системы счисления в десятичную

```python
x = input()
length = len(x) - 1
y = 0
for digit in x:
    y += int(digit) * 2 ** length
    length -= 1
print(y)
```

### 2. Вычисление дроби

```python
from fractions import Fraction


a_n, a_d = map(int, input().split())
b_n, b_d = map(int, input().split())
c_n, c_d = map(int, input().split())
a = Fraction(a_n, a_d)
b = Fraction(b_n, b_d) 
c = Fraction(c_n, c_d) 
formula = float((b / a) + (b / (a + c)) - (c / (c - a)))
print(f"{formula:0.4f}")
```

### 3. Сумма последовательности чисел

```python
count = 0
sum = 0
digit =  int(input())
while digit != 0:
    count += 1
    sum += digit
    digit =  int(input())
print(count, sum)
```

### 4. Элемент геометрической прогрессии

```python
b1, q, a = map(float, input().split())
n = 1
while b1 * q ** (n - 1) < a:
    n += 1
print(n)
```

### 5. Банковский вклад*

```python
amount, period, rate = map(float, input().split())
months = int(period * 12)
rate_month = rate / 12 / 100
with open('output.csv', 'w', encoding='utf-8') as csv_file:
    print('Месяц', 'Сумма на вкладе', 'Начисление', sep=',', file=csv_file)
    for i in range(1, months + 1):
        deposit_amount = amount * (1 + rate_month)
        accrual = deposit_amount - amount
        print(i, f'{deposit_amount:0.2f}', f'{accrual:0.2f}', sep=',', file=csv_file)
        amount = deposit_amount
```
