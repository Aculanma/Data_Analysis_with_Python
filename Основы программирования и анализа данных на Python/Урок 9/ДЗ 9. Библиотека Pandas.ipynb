### 2. Ошибка выжившего

```python
import pandas as pd
import numpy as np


df = pd.read_csv('input.csv', sep=',', header=None)
lst = []
for num in range(df.shape[1]):
    lst.append(np.mean(df[num]))
index = lst.index(min(lst))
print(index+1)
```

### 3. Электрик Станислав и интернет-казино

### 4. Инвестор Генадий и новый Мерседес

```python
import pandas as pd
import numpy as np


df = pd.read_csv('input.csv', sep=',')
if df.sum().sum() >= 8 * 10 ** 6:
    print(1, df.sum().idxmax())
else:
    print(0, df.sum().idxmax())
```
