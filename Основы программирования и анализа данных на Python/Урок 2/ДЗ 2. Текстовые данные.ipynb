### 1. Количество символов в строке

Задана строка *s*, состоящая из символов. Необходимо посчитать их количество ∣*s*∣.

```python
text = input()
print(len(text))
```

### 2. Строка в обратном порядке

Измените порядок символов в строке *s* на обратный.

```python
text = input()
print(text[::-1])
```

### 3. Какой логин?

```python
text = input()
print(text[:text.find('@')])
```

### 4. Подстрока в двойных кавычках

```python
text = input()
print(text[text.find('"')+1:text.rfind('"')])
```

### 5. Поменялись!

```python
text = input() 
print(text[text.find(' ')+1:] + ' ' + text[:text.find(' ')])
```

### 6. Как? Как угодно!

```python
text = input()
print(text.replace('(', '').replace(')','').replace(' ','').replace('-',''))
```

### 7. Щйхс Чёибса

```python
text = input()
new = ''
for i in text:
    if i.isalpha():
        new += chr(ord(i)+1)
    else:
        new += i
print(new)
```
